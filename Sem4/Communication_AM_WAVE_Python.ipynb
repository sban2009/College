```
#PRESS CTRL+F9 TO RUN PROJECT CELL BY CELL
#THEN SCROLL DOWN TO LAST CELL TO INPUT USER DATA

import numpy as np
from math import pi
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc('figure', figsize=(20, 10))

#TIME AXIS CREATION

Fs=2000; #sampling frequency in Hz
t=np.arange(0,0.75,1/Fs)

#CARRIER WAVE GENERATION
Fc=50
Ac=1
c=Ac*np.cos(2*pi*Fc*t)

plt.plot(t,c)
plt.title('CARRIER')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude(V)')
plt.grid(True)
```

```
#MESSAGE WAVE GENERATION
Fm=2
Am=0.5
m=Am*np.sin(2*pi*Fm*t)

plt.plot(t,m)
plt.title('MESSAGE')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude(V)')
plt.grid(True)
```

```
#AM WAVE GENERATION
s=c*(1+m/Ac)

#PLOT 1
plt.plot(t,s)
plt.title('UNDERMODULATED')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude(V)')
plt.grid(True)
```

```
#MESSAGE WAVE CHANGES
Am=1

#REGENERATION OF MESSAGE AND AM WAVES
m=Am*np.sin(2*pi*Fm*t)

s=c*(1+m/Ac)

#PLOT 2
plt.plot(t,s)
plt.title('CRITICALLY MODULATED')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude(V)')
plt.grid(True)
```

```
#MESSAGE WAVE CHANGES
Am=2.5

#REGENERATION OF MESSAGE AND AM WAVES
m=Am*np.sin(2*pi*Fm*t)

s=c*(1+m/Ac)

#PLOT 3
plt.plot(t,s)
plt.title('OVERMODULATED')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude(V)')
plt.grid(True)
```

```
#GENERALISED

#TIME AXIS CREATION
Fs=float(input("Enter sampling frequency (in Hz): ")) #sampling  frequency in Hz
t1,t2=str.split(input("Enter time interval in format t1,t2 (in s): "),',')
t1=float(t1)
t2=float(t2)
t=np.arange(t1,t2,1/Fs)

#CARRIER WAVE GENERATION
Fc=float(input("Enter carrier frequency (in Hz): "))
Ac=float(input("Enter carrier amplitude (in V): "))
            
c=Ac*np.cos(2*pi*Fc*t)

#MESSAGE WAVE GENERATION
Fm=float(input("Enter message frequency (in Hz): "))
Am=float(input("Enter message amplitude(in V): "))
            
m=Am*np.sin(2*pi*Fm*t)

#AM WAVE GENERATION
s=c*(1+m/Ac)

#TYPE OF MODULATION
modIndex=Am/Ac
type=''
if modIndex<1:
    type="UNDERMODULATED"
elif modIndex>1:
    type="OVERMODULATED"
else:
    type="CRITICALLY MODULATED"
    
#PLOT DEFAULT
plt.plot(t,s)
plt.title(type)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.savefig('AM_Wave.png') #SAVES TO MACHINE

#PRESS CTRL+F9 TO RUN CELL
```
