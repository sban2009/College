```
'''
MONTY HALL PROBLEM
CSE 2ND YEAR
GROUP 1
M401 MICRO PROJECT
'''
import random
doors=["","",""] #LIST OF DOORS
goatdoor=[]      #LIST OF DOORS NOT CONTAINING PRIZE
swap=0           #NO. OF SWAPPED ATTEMPTS
Xswap=0          #NO. OF RETAINED ATTEMPTS
swapWins=0       #WINS AFTER SWAPPING
XswapWins=0      #WINS AFTER RETAINING (NOT SWAPPING)
exit=False
while(not exit):
    x=random.randint(0,2)      #DOOR CONTAINING PRIZE
    doors[x]="PRIZE"           #INDEX x OF doors HAS PRIZE
    for i in range(0,3):       #FORMING THE LIST goatdoor
        if i!=x:
            doors[i]="GOAT"     #VALUES IN doors NOT EQUAL TO "PRIZE"
            goatdoor.append(i)  #ARE MADE "GOAT"
    ch1=int(input("Enter choice of door (1/2/3): ")) #USER INPUT OF DOOR CHOICE
    ch1-=1 #SINCE WE ARE WORKING WITH START INDEX 0 (SO 1/2/3 => 0/1/2)
    opendoor=random.choice(goatdoor) #GIVES US RANDOM INDEX FROM goatdoor
    while opendoor==ch1:
        opendoor=random.choice(goatdoor)   #opendoor NOT EQUAL TO CHOICE ch1 OF USER
    print("Door%2d has GOAT"%(opendoor+1)) #USER IS SHOWN WHAT'S BEHIND ONE OF THE REMAINING DOORS
    ch2=input("Do you want to swap? (Y/N): ")    #PROMPT TO SWAP
    W="PLAYER WINS"  #STRINGS TO STORE
    L="PLAYER LOSES" #WIN/LOSS MESSAGES
    if str.upper(ch2)=="Y":                    #SWAPS
        swap+=1
        if doors[ch1]=="GOAT":                 #INITIAL CHOICE GOAT: USER WINS
            print(W)
            swapWins+=1                        ##WIN BY SWAPPING
        else:                                  #OR ELSE USER LOSES
            print(L)
    else:                                      #RETAINS
        Xswap+=1
        if doors[ch1]=="GOAT":                 #INITIAL CHOICE GOAT: USER LOSES
            print(L)
        else:                                  #OR ELSE USER WINS
            print(W)
            XswapWins+=1                       ##WIN BY RETAINING
    if str.upper(input("Exit? (Y/N): "))=="Y":
        exit=True                              #EXIT CONDITION
print("\n")
print("Swap attempts:",swap)
print("Wins after swapping: ",swapWins)
print("\n")
print("Retain attempts:",Xswap)
print("Wins after retaining: ",XswapWins)
```
